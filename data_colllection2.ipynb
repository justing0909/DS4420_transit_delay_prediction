```python
# Function to create a unique ID for each stop
def create_stop_identifier(stop_data):
    return f"{stop_data['route']}{stop_data['stop_id']}_{stop_data['direction_id']}"

# NEW: Function to fetch predictions for a given stop ID
def fetch_predictions(stop_id, api_key):
    url = f"https://api-v3.mbta.com/predictions?filter[stop]={stop_id}"
    params = {
        "api_key": api_key,
        "page[limit]": 700
    }
    
    response = requests.get(url, params=params)
    if response.status_code == 200:
        predictions_data = response.json()
        
        # Extract relevant information from the predictions
        predictions = []
        for prediction in predictions_data.get('data', []):
            attributes = prediction.get('attributes', {})
            predictions.append({
                'arrival_time': attributes.get('arrival_time'),
                'departure_time': attributes.get('departure_time'),
                'departure_uncertainty': attributes.get('departure_uncertainty'),
                'direction_id': attributes.get('direction_id'),
                'status': attributes.get('status'),
                'stop_sequence': attributes.get('stop_sequence'),
                'schedule_relationship': attributes.get('schedule_relationship')
            })
        
        return predictions
    else:
        print(f"Error fetching predictions for stop {stop_id}: {response.status_code}")
        return []

# Function to fetch schedules for a given stop ID
def fetch_schedules(stop_id, api_key):
    url = f"https://api-v3.mbta.com/schedules?filter[stop]={stop_id}"
    params = {
        "api_key": api_key,
        "page[limit]": 700
    }
    
    response = requests.get(url, params=params)
    if response.status_code == 200:
        schedules_data = response.json()
        
        # Extract relevant information from the schedules
        schedules = []
        for schedule in schedules_data.get('data', []):
            attributes = schedule.get('attributes', {})
            schedules.append({
                'arrival_time': attributes.get('arrival_time'),
                'departure_time': attributes.get('departure_time'),
                'departure_uncertainty': attributes.get('departure_uncertainty'),
                'direction_id': attributes.get('direction_id'),
                'status': attributes.get('status'),
                'stop_sequence': attributes.get('stop_sequence'),
                'schedule_relationship': attributes.get('schedule_relationship')
            })
        
        return schedules
    else:
        print(f"Error fetching schedules for stop {stop_id}: {response.status_code}")
        return []



# # NEW: Optional: Save the result to a CSV file
# if result_array is not None and result_array.size > 0:
#     df = pd.DataFrame(result_array, columns=['stop_id', 'bias', 'precipitation', 'temperature', 'delay', 'scheduled_time', 'actual_time'])
#     df.to_csv('mbta_weather_delay_data.csv', index=False)
#     print("Data saved to mbta_weather_delay_data.csv")
```

```python

# NEW: Function to fetch weather data
def fetch_weather_data():
    url = "https://api.open-meteo.com/v1/forecast?latitude=42.3584&longitude=-71.0598&hourly=temperature_2m,precipitation,rain,showers,snowfall,snow_depth&minutely_15=temperature_2m,snowfall,precipitation,rain&past_days=14&forecast_days=16&wind_speed_unit=mph&temperature_unit=fahrenheit&precipitation_unit=inch"
    
    response = requests.get(url)
    if response.status_code == 200:
        weather_data = response.json()
        
        # Restructure the weather data
        weather = []
        for i, time in enumerate(weather_data['minutely_15']['time']):
            weather.append({
                'time': time,
                'temperature_2m': weather_data['minutely_15']['temperature_2m'][i],
                'precipitation': weather_data['minutely_15']['precipitation'][i],
            })
        
        # Change dictionaries to rows
        weather_array = np.array([list(d.values()) for d in weather])
        
        # Add bias
        weather_array = np.insert(weather_array, 0, 1, axis=1)
        
        return weather_array
    else:
        print(f"Error fetching weather data: {response.status_code}")
        return np.array([])
```

```python
# MODIFIED: Added error handling and docstring
def create_time_difference_array_within_tolerance(predictions, schedules):
    time_diff_array = []
    for pred in predictions:
        arrival_time = pred['arrival_time']
        if arrival_time is None:  # NEW: Check for None
            continue
        
        arrival_time_dt = datetime.fromisoformat(arrival_time.replace('Z', '+00:00'))
        for schedule in schedules:
            schedule_arrival_time = schedule['arrival_time']
            if schedule_arrival_time is None:  # NEW: Check for None
                continue
                
            schedule_arrival_time_dt = datetime.fromisoformat(schedule_arrival_time.replace('Z', '+00:00'))
            # Check if the times are within +/- 3 minutes
            if abs((arrival_time_dt - schedule_arrival_time_dt).total_seconds()) <= 180:
                time_difference = (arrival_time_dt - schedule_arrival_time_dt).total_seconds()
                time_diff_array.append([schedule_arrival_time, time_difference])
                break  # Stop checking once a match is found
    
    return np.array(time_diff_array) if time_diff_array else np.array([])
```

```python
# MODIFIED: Added stop_identifier parameter and modified the data structure
def match_weather_to_time_diff(weather_array, time_diff_array, stop_identifier):  # NEW: Added stop_identifier parameter
    matched_data = []
    for i, weather in enumerate(weather_array):
        weather_time = datetime.fromisoformat(weather[1].replace('Z', '+00:00')).astimezone(None)
        for j, time_diff in enumerate(time_diff_array):
            schedule_arrival_time = datetime.fromisoformat(time_diff[0].replace('Z', '+00:00')).astimezone(None)
            # Calculate actual time by adding delay to scheduled time
            delay_seconds = float(time_diff[1])
            actual_arrival_time = schedule_arrival_time + timedelta(seconds=abs(delay_seconds))
            
            if abs((schedule_arrival_time - weather_time).total_seconds()) <= 900:
                matched_data.append({
                    'stop_id': stop_identifier,           # NEW: Added stop identifier
                    'bias': weather[0],                   # Bias from weather data
                    'precip': weather[3],                 # Precipitation from weather data
                    'temp': weather[2],                   # Temperature from weather data
                    'delay': time_diff[1],                # Time difference (delay)
                    # 'scheduled_time': time_diff[0],       # Scheduled arrival time
                    # 'actual_time': actual_arrival_time.isoformat()  # Calculated actual arrival time
                })
                break  # Stop checking once a match is found
    return matched_data
```

```python
# MODIFIED: Added stop_id to the output array
def create_matched_weather_array(matched_weather_data):
    matched_weather_array = []
    for data in matched_weather_data:
        matched_weather_array.append([
            data['stop_id'],         # NEW: Added stop identifier
            data['bias'],            # Bias
            data['precip'],          # Precipitation
            data['temp'],            # Temperature
            data['delay'],           # Delay (time difference)
            # data['scheduled_time'],  # Scheduled arrival time
            # data['actual_time']      # Actual arrival time
        ])
    return np.array(matched_weather_array)
```

```python
# NEW: Main function to process all stops
def process_all_stops(api_key):
    # Fetch weather data once (same for all stops)
    weather_array = fetch_weather_data()
    if weather_array.size == 0:
        print("Failed to fetch weather data.")
        return None
    
    all_matched_data = []
    
    for stop_info in stops_data:
        stop_id = stop_info["stop_id"]
        stop_identifier = create_stop_identifier(stop_info)
        print(f"Processing stop {stop_identifier} ({stop_info['name']})...")
        
        # Fetch predictions and schedules for this stop
        predictions = fetch_predictions(stop_id, api_key)
        schedules = fetch_schedules(stop_id, api_key)
        
        if not predictions or not schedules:
            print(f"No predictions or schedules found for stop {stop_identifier}.")
            continue
        
        # Create time difference array
        time_diff_array = create_time_difference_array_within_tolerance(predictions, schedules)
        if time_diff_array.size == 0:
            print(f"No time differences found for stop {stop_identifier}.")
            continue
        
        # Match weather data to time differences
        matched_weather_data = match_weather_to_time_diff(weather_array, time_diff_array, stop_identifier)
        if not matched_weather_data:
            print(f"No matched weather data found for stop {stop_identifier}.")
            continue
        
        # Add to our overall dataset
        all_matched_data.extend(matched_weather_data)
        
        # Sleep briefly to avoid hitting API rate limits
        time.sleep(1)
    
    # Create the final array from all matched data
    if all_matched_data:
        final_array = create_matched_weather_array(all_matched_data)
        print("Final matched weather array:")
        print(final_array)
        return final_array
    else:
        print("No matched data found for any stops.")
        return None
```
